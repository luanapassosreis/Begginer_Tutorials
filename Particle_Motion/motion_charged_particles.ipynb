# Motion of a charge particle in a slowly changing magnetic field

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
q = 1.0
m = 1.0
dt = 1e-3
t0 = 0
t1 = 10
t = np.linspace(t0, t1, int((t1 - t0)/dt))
n = len(t)
print("Number of elements in t array: ", n)

r = np.zeros((n, 3))
v = np.zeros((n, 3))
```

```python
## initial conditions
r[0] = [0.0, 0.0, 0.0]
v[0] = [2.0, 0.0, 3.0]
```

```python
## magnetic field
B = np.zeros((n, 3))
B[0] = np.array([0.0, 0.0, 5.0])
dB = np.array([0.0, 0.0, 1e-3])

for i in range(len(t) - 1):
    a = q / m * np.cross(v[i], B[i])
    v[i + 1] = v[i] + a * dt
    r[i + 1] = r[i] + v[i + 1] * dt
    B[i + 1] = B[i] + dB
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot(r[:, 0], r[:, 1], r[:, 2])
ax.set_xlabel("$x$", fontsize=26)
ax.set_ylabel("$y$", fontsize=26)
ax.set_zlabel("$z$", fontsize=26)
plt.show()
```

# Trying to plot Bremsstrahlung emission

```python
## constants
mu_0 = 4 * np.pi * 1e-7  # [H m-1] vacuum permeability
q = 1.6e-19              # [Coulomb] electron
c = 3e8                  # [m s-1]
a = 1e16                 # [m s-2] acceleration (example)
beta = 0.9               # v/c relativistic
```

```python
## angles for plotting
theta = np.linspace(0, np.pi, 1000)

## differential power distribution formula
dP_dOmega = (mu_0 * q**2 * a**2 * np.sin(theta)**2) / (16 * np.pi**2 * c * (1 - beta * np.cos(theta))**5)
```

```python
## normalize for plotting
dP_dOmega /= np.max(dP_dOmega)
```

```python
## plot in polar coordinates
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, polar=True)
ax.plot(theta, dP_dOmega, label='Bremsstrahlung Radiation Pattern')
```

```python
## convert to Cartesian coordinates for x-z plane plot
x = dP_dOmega * np.sin(theta)
z = dP_dOmega * np.cos(theta)
```

```python
## cartesian plot
fig, ax = plt.subplots(figsize=(12, 10))
ax.plot(x, z, label='Bremsstrahlung Radiation Pattern')
ax.set_xlabel("$x$", fontsize=26)
ax.set_ylabel("$z$", fontsize=26)
ax.set_title('Bremsstrahlung Radiation Pattern in x-z Plane')
ax.legend()
plt.show()
```

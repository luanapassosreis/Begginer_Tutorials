Dois conceitos importantes:

# <font color='#E9AB17'>***Zero Age Main Sequence (ZAMS)***</font>

* Início da sequência principal na trajetória evolutiva de estrela

# <font color='#E9AB17'>***Terminal Age Main Sequence (TAMS)***</font>

* Fim da sequência principal e entrada no ramo das gigantes (evolução pós-sequência principal)

Durante as aulas, trabalhamos os estágios "clássicos" da evolução estelar. Portanto, sabemos que durante a sequência principal, as estrelas queimam hidrogênio em seus núcleos (via, por exemplo, cadeia próton-próton e ciclo CNO). Sendo assim, o fim da sequência principal ocorre quando não há mais disponibilidade de hidrogênio no núcleo para queima.

Portanto, chegamos ao desafio: <font color='#E9AB17'>***como definir ZAMS e TAMS?***</font>

```python
# PASSO 6
# nessa célula do Colab, iremos acessar a pasta "data_models" novamente. Aqui, encontram-se todos os modelos do MESA produzidos para a aula.
# caso queira testar os códigos fornecidos aqui com outros modelos, vc pode criar os seus próprios via o link do MESA web

%cd
!pwd
%cd /content/gdrive/My\ Drive/MESA_AGA0100/data_models/1M_Zsun_complete

# agora, trabalharemos dentro desse diretório ao longo de todo resto da atividade.
# Portanto, se vc chegou até aqui, não tem mais onde dar errado

# esses arquivos "history.data" contém os parâmetros físicos do modelo ao longo do tempo de simulação
# por conveniência, esses modelos (também do Sol) foram produzidos já separados nos diferentes estágios evolutivos

h1 = mr.MesaData('LOGS_to_end_core_h_burn/history.data') # seq. principal
h2 = mr.MesaData('LOGS_to_start_he_core_flash/history.data') # início do ramo das gigantes vermelhas (RGB)
h3 = mr.MesaData('LOGS_to_end_core_he_burn/history.data') # ramo horizontal
h4 = mr.MesaData('LOGS_to_end_agb/history.data') # ramo assintótico (AGB)
h5 = mr.MesaData('LOGS_to_wd/history.data') # pós AGB para anã branca


# os comando abaixo simplesmente coloca em uma variável o diretório onde está esse arquivo (vcs podem vê-lo nas pastas do material)
# isso serve para o que vem (também comentado) logo abaixo
#dir = mr.MesaLogDir('./LOGS_to_end_core_h_burn')

# também podemos colocar em uma variável o perfil radial / estrutura interna da estrela (propriedades físicas em função do raio)
# essa linha de comando abaixo simplesmente pega o último perfil registrado na simulação
# não utilizaremos isso em aula, mas fica aqui para quem quiser.
#p1 = dir.profile_data()
```

```python
# PASSO 7
# DEFININDO ZAMS E TAMS

# para entrar na sequência principal, a fonte dominante de energia da estrela é a queima do hidrogênio.
# quando a LUMINOSIDADE associada à queima de H é responsável por quase (99%) toda energia liberada, temos a chegada em ZAMS

# para sair da sequência principal, quase todo hidrogênio do núcleo deve ter sido consumido
# quando a FRAÇÃO DE MASSA do hidrogênio no núcleo for quase (0.0015) nula, temos TAMS


# esse é um pequena script para obter as informações que queremos através de um "loop"
# basicamente, visitamos todas as células da tabela na qual os dados estão e avaliamos qual delas pasa no critério
for i in range(len(h1.log_L)):

  if 10**(h1.log_LH[i]) / 10**(h1.log_L[i]) >= 0.99: # aqui aplica-se o critério
    print("Razão entre luminosidade fornecida pela queima do H e luminosidade total em ZAMS: "+str(round( 10**(h1.log_L[i]) / 10**(h1.log_LH[i]) ,5)))
    print("Idade ZAMS: "+str("{:e}".format(round(h1.star_age[i],5))))
    print("Temperatura efetiva ZAMS: "+str(round(h1.Teff[i], 1))+" K")
    print("Luminosidade ZAMS: "+str(round( 10**(h1.log_L[i]) , 5) )+" Lsol")

    ZAMS_Teff = h1.log_Teff[i]
    ZAMS_L = h1.log_L[i]
    ZAMS_age = h1.star_age[i]
    break

print() # pula linha

# esse script abaixo faz as mesmas coisas que o de cima, mas para TAMS
for i in range(len(h1.model_number)):
  if h1.center_h1[i]/h1.center_h1[0] <= 0.0015: # aqui aplica-se o critério

    # vamos colocar na tela essas informações para discutir
    print("Fração inicial de massa de H no núcleo: "+str(round(h1.center_h1[0],4)))
    print("Idade TAMS: "+str("{:e}".format(round(h1.star_age[i], 1)))+" anos")
    print("Fração de massa de H no núcleo em TAMS: "+str(round(h1.center_h1[i],5)))
    print("Razão entre fração de massa de H e seu valor inicial: "+str(round(h1.center_h1[i]/h1.center_h1[0],10)))
    print("Temperatura efetiva TAMS: "+str(round(h1.Teff[i], 1))+" K")
    print("Luminosidade TAMS: "+str(round( 10**(h1.log_L[i]) , 5) )+" Lsol")

    # ao mesmo tempo, vou salvando os valores dos parâmetros estelares em TAMS
    TAMS_Teff = h1.log_Teff[i]
    TAMS_L = h1.log_L[i]
    TAMS_age = h1.star_age[i]
    break

# mesma coisa que os anteriores, só que, agora, para a idade atual do Sol
# isso pode ser estimado de forma independente, por exemplo, com datação de minérios
for i in range(len(h1.star_age)):
  if h1.star_age[i] > Age_sun:
    dif1 = h1.star_age[i-1] - Age_sun
    dif2 = h1.star_age[i] - Age_sun

    if abs(dif1) < abs(dif2) == True:
      sun_Teff = h1.log_Teff[i-1]
      sun_L = h1.log_L[i-1]
      sun_age = h1.star_age[i-1]
      break
    else:
      sun_Teff = h1.log_Teff[i]
      sun_L = h1.log_L[i]
      sun_age = h1.star_age[i]
      break
```

# <font color='#E9AB17'>**Relação massa$-$luminosidade**</font>


$L \propto M^{\alpha} \Rightarrow \dfrac{L}{L_{\odot}} \cong \left( \dfrac{M}{M_{\odot}} \right)^\alpha$

onde $\alpha$ (3.5$-$4.0) é uma constante calibrada empiricamente.

Podemos utilizar os valores obtidos com as trajetórias evolutivas calculadas teoricamente com a biblioteca MESA para determinar o parâmetro $\alpha$!

Para isso, vamos considerar essa relação apenas em ZAMS. Podemos escrever algo como:


$\dfrac{L_{\rm ZAMS}}{L_{\odot \rm ZAMS}} \cong \left( \dfrac{M_{\rm ZAMS}}{M_{\odot \rm ZAMS}} \right)^\alpha$


Para 1 $M_{\odot}$, temos $L_{\rm ZAMS} = 0.69 \ L_{\odot}$. Para 0.5 $M_{\odot}$, temos $L_{\rm ZAMS} = 0.04 \ L_{\odot}$. Ficamos com:

$\dfrac{0.04 L_{\odot} }{0.69 L_{\odot}} \cong \left( \dfrac{0.5M_{\odot}}{1.0 M_{\odot}} \right)^\alpha \Rightarrow \log{ \left( \dfrac{0.04}{0.69} \right) } \cong \alpha \log{0.5} \Rightarrow \alpha \cong 4.1 $

Em aula, utilizamos a aproximação $\alpha \approx 4$, o que é razoável. Agora, já sabemos que, como as estrelas se deslocam um pouco na sequência principal, essa relação empírica tem certa limitação.

**Atenção:** o que acontece caso consideremos a relação massa$-$luminosidade em TAMS? Perceba que a variação de luminosidade ao longo do tempo de vida na sequência principal da estrela de massa 0.5 $M_{\odot}$ é maior que para 1.0 $M_{\odot}$, logo o parâmetro $\alpha$ deve, também, ser diferente. Tomara que seja mais próxio de $\sim$3, já que, em aula, dissemos que esse valor é encontrado na literatura entre 3.5$-$4.0.

# <font color='#E9AB17'>**Tempo de vida na sequência principal**</font>


$\dfrac{t_{\rm SP}}{t_{\odot}} \cong \left( \dfrac{M_{\odot}}{M} \right)^{\alpha -1} $, onde acabamos obter $\alpha \approx 4.1$.

Que tal calcularmos os tempos de vida na sequência principal e comparar com aqueles obtidos pelo MESA? Tomara que dê certo, já que estamos "calibrando" a relação com os próprios modelos do MESA.

Tempo de vida na SP para 1 $M_{\odot}$: $t_{\odot} \approx 9.05$ Gano.

Vamos testar para $t_{\rm SP}$ da estrela de 0.5 $M_{\odot}$.

$\dfrac{t_{\rm SP}(0.5 M_{\odot})}{9.05 \times 10^9 \ \rm anos} \cong \left( \dfrac{1.0 M_{\odot}}{0.5 M_{\odot}} \right)^{4.1 -1} \Rightarrow t_{\rm SP}(0.5 M_{\odot}) \approx 0.77 \times 10^{11}$ anos.

Valor esperado (calculado a partir do modelo do MESA): $1.20 \times 10^{11}$ anos.

**Spoiler:** para 5 $M_{\odot}$, essa relação funciona melhor.

Portanto, não existe nenhuma estrela de massa $\leq$ 0.5 $M_{\odot}$ que já tenha saído da sequência principal.
